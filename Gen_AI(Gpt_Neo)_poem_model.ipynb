```python
from transformers import pipeline

generator = pipeline('text-generation', model="EleutherAI/gpt-neo-1.3B") # load model
```

```python
user_input = input("Enter a theme for your poem: ") # take user prompt
promt = (f"A whimsical poem about {user_input}")
```

```python
# generate poem
result = generator(promt, max_length=100, num_return_sequences=5, truncation=True, temperature=0.7, top_p=0.9)
print(result[0]['generated_text'])
```
